This is based on inclusion and exclusion concept 
Seven people are there with 4 choice of season so possibility = 4^7
These are the total outcomes 
Now outcomes with three distinct season = 4C3*3^7 but in this the outcomnes with two seasons are double counted 
Now outcomes with two distinct season = 4C2*2^7 is subtracted out
so outcomes wuth one seasons = 4C1*1^7 
Favourable outcomes = 4^7-(4C3*3^7-4C2*2^7+4C1*1^7 )= 8400
Total Outcomes = 4^7

Required probability = 8400/4^7 =0.51269
